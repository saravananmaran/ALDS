```python
import numpy as np
import pandas as pd
df = pd.read_csv('/amazonreviews.tsv', sep='\t')
df.head()
```

```python
df['label'].value_counts()
```

```python
df.dropna(inplace=True)
```

```python
df['label'].value_counts()
```

```python
blanks = []
for i,lb,rv in df.itertuples():
  if type(rv)==str:
    if rv.isspace():
      blanks.append(i)

df.drop(blanks, inplace= True)
df['label'].value_counts()
```

```python
sid.polarity_scores(df.iloc[0]['review'])
df.loc[0]['label']
```

```python
df['scores'] = df['review'].apply(lambda review: sid.polarity_scores(review))
df.head()
```

```python
sid.polarity_scores(df.loc[0]['review'])
df.loc[0]['scores']
```

```python
df['comp_score'] = df['scores'].apply(lambda d:d['compound'])
df.head()
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

df['pred_label'] = df['comp_score'].apply(lambda score: 'pos' if score >=0 else 'neg')

print(accuracy_score(df['label'],df['pred_label']))
print(classification_report(df['label'],df['pred_label']))
print(confusion_matrix(df['label'],df['pred_label']))
```

```python
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
type(iris)
```

```python
import pandas as pd
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
print(iris.DESCR)
```

```python
x = iris.data
```

```python
x
```

```python
y = iris.target
```

```python
y
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
y.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar_object = MinMaxScaler()
```

```python
scalar_object.fit(x_train)
```

```python
scalar_x_train = scalar_object.transform(x_train)
```

```python
scalar_x_test = scalar_object.transform(x_test)
```

```python
x_train.max()
```

```python
scalar_x_train.min()
```

```python
x_train
```

##BUILDING THE NETWORK WITH KERAS


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import tensorflow as tf
```

```python
model = Sequential()
model.add(Dense(8, input_dim=4, activation='relu'))
model.add(Dense(8, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

###FIT (Train) the Model

```python
model.fit(scalar_x_train, y_train, epochs=150, verbose=2)
```

###

```python
model.predict(scalar_x_test)
```

```python
model.metrics_names
```

```python
model.evaluate(scalar_x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
predicted_probabilities = model.predict(scalar_x_test)
predictions = np.argmax(predicted_probabilities, axis=1)
```

```python
y_test.argmax(axis=1)
```

```python
confusion_matrix(y_test.argmax(axis=1), predictions)
```

```python
print(classification_report(y_test.argmax(axis=1), predictions))
```

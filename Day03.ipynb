```python
import numpy as np
import pandas as pd
df = pd.read_csv('/amazonreviews.tsv', sep='\t')
df.head()
```

```python
df['label'].value_counts()
```

```python
df.dropna(inplace=True)
```

```python
df['label'].value_counts()
```

```python
blanks = []
for i,lb,rv in df.itertuples():
  if type(rv)==str:
    if rv.isspace():
      blanks.append(i)

df.drop(blanks, inplace= True)
df['label'].value_counts()
```

```python
sid.polarity_scores(df.iloc[0]['review'])
df.loc[0]['label']
```

```python
df['scores'] = df['review'].apply(lambda review: sid.polarity_scores(review))
df.head()
```

```python
sid.polarity_scores(df.loc[0]['review'])
df.loc[0]['scores']
```

```python
df['comp_score'] = df['scores'].apply(lambda d:d['compound'])
df.head()
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

df['pred_label'] = df['comp_score'].apply(lambda score: 'pos' if score >=0 else 'neg')

print(accuracy_score(df['label'],df['pred_label']))
print(classification_report(df['label'],df['pred_label']))
print(confusion_matrix(df['label'],df['pred_label']))
```
